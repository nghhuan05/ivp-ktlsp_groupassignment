```python
import pandas as pd

# Tên file
file_name = "marketing_campaign.csv"

# Rất quan trọng: phải thêm sep='\t' (Tab-separated)
df_marketing = pd.read_csv(file_name, sep='\t')

# Xem 5 dòng đầu tiên
print("--- 5 dòng đầu của dữ liệu Marketing ---")
print(df_marketing.head())
```

```python
# Xem thông tin và kiểu dữ liệu
print("\n--- Thông tin dữ liệu (Info) ---")
df_marketing.info()

# Xem các cột có dữ liệu thiếu (Missing Data)
print("\n--- Kiểm tra dữ liệu thiếu (NaN) ---")
print(df_marketing.isnull().sum())
```

```python
from autoviz.AutoViz_Class import AutoViz_Class

# 1. Cài đặt (chỉ chạy 1 lần nếu chưa cài)
# !pip install autoviz

# 2. Khởi tạo AutoViz
AV = AutoViz_Class()

# 3. Chạy AutoViz để tạo báo cáo
# depVar = 'Response' -> báo cho AutoViz đây là biến mục tiêu
print("Đang chạy AutoViz... quá trình này sẽ tạo nhiều biểu đồ...")
df_av = AV.AutoViz(
    filename="", 
    depVar='Response', 
    dfte=df_marketing, 
    header=0, 
    verbose=2,
    chart_format='html' 
)
```
